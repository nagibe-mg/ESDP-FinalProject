```python
from datetime import datetime, timedelta
import os
from RealDailyRoutineERA5 import compare_years_batch_mode#, animation_years


start_date = datetime(2014, 5, 1)
end_date = datetime(2014, 5, 31)

path_2014 = "era5_precipitation.zarr"
path_2015 = "era5_precipitation2015.zarr"
diff_store = "diff_2014_vs_2015.zarr"


img_folder = "frames_interannual"
os.makedirs(img_folder, exist_ok=True)


current = start_date
while current <= end_date:
    d1_str = current.strftime("%Y-%m-%d")
    d2_obj = current.replace(year=2015)
    d2_str = d2_obj.strftime("%Y-%m-%d")
    
    # Run Function
    compare_years_batch_mode(
        path1=path_2014,
        path2=path_2015,
        diff_store_path=diff_store,
        date1=d1_str,
        date2=d2_str,
        output_dir=img_folder
    )
    
    current += timedelta(days=1)

print("Monthly comparison complete!")
print(f"Images saved in: {img_folder}/")
print(f"Data saved in: {diff_store}")
```

```python
import imageio.v2 as imageio
import os
import glob

def create_gif(image_folder, output_gif_name):
    print(f"scanning folder: {image_folder}")
    
    images = sorted(glob.glob(os.path.join(image_folder, "*.png")))
    #images = glob.glob(os.path.join(image_folder, "*.png"))
    print(f"   Found {len(images)} frames.")
    frames = []
    for filename in images:
        frames.append(imageio.imread(filename))
    
    print(f"Saving GIF to {output_gif_name}...")
    imageio.mimsave(output_gif_name, frames, fps=1)
    #imageio.mimsave(gif_path, images, fps=1) # 1 frame per second
    print("Done!")

folder_name = "frames_interannual" 
output_file = "precipitation_comparison_2014_vs_2015.gif"

create_gif(folder_name, output_file)
```
