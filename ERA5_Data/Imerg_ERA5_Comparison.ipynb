```python
import xarray as xr
import numpy as np

def check_time_comp(store_a, store_b):
    ds_a = xr.open_zarr(store_a, group="NSIDE_128", consolidated=True)
    ds_b = xr.open_zarr(store_b, group="NSIDE_128", consolidated=True)
    
    t_a = ds_a.time.values[0]
    t_b = ds_b.time.values[0]
    
    print(f"Store A First Time: {t_a} (Type: {type(t_a)})")
    print(f"Store B First Time: {t_b} (Type: {type(t_b)})")
    
    # Check if they are "Naive" (no Z/UTC attached) or "Aware"
    # Numpy datetime64 is usually naive (just a raw number).
    # If they both look like '2014-05-01T00:00:00.000000000', you are safe.
    
    if t_a == t_b:
        print("Timestamps match perfectly.")
    else:
        print("Timestamps do not match.")


```

```python
check_time_comp("era5_precipitation.zarr", "era5_precipitation.zarr")
```
