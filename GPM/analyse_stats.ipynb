```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
daily_stats = pd.read_csv('daily_precip_stats.csv', delimiter=',')
daily_stats
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(daily_stats['date'], daily_stats['max_mm'], marker='o', label='maximum')
ax.plot(daily_stats['date'], daily_stats['mean_mm'], marker='o', label='mean')
ax.plot(daily_stats['date'], daily_stats['min_mm'], marker='o', label='minimum')

ax.set_xlabel('Date')
ax.set_ylabel('Precipitation [mm/day]')

plt.xticks(rotation=90)
plt.legend()

plt.tight_layout()
plt.savefig('daily_precip_stats.png', dpi=300)
```
